### Tratamento de variáveis categóricas

 - O que são variáveis categóricas, discretas e contínuas
 - Porque tratar variáveis categóricas
 - Criando variáveis Dummy
 - Indexando variáveis categóricas

#### 3 Tipos de variáveis categóricas:
    
    - Categóricas: São variáveis usadas para categorizar os dados, não necessariamente seguem uma ordem deﬁnida.
    Exemplo: Cores de Carros
        
     - Discretas: São variáveis numéricas que tem um número ﬁnito que pode ser contado
    Exemplo: Número de reclamações recebidas
        
     - Contínuas: Variáveis numéricas com um número inﬁnito de possibilidades entre dois valores
    Exemplo: Medida de um trajeto

#### Existem duas técnicas mais comuns para realizar este tratamento:

 - Criar variáveis Dummy
 - Indexar variáveis categóricas

## Aplicando o Get Dummies

 - Variáveis Dummy, são quebras da variável principal, que podem ser representadas 
pelos números 0 e 1 (0 para falso, e 1 para verdadeiro).

 - No Python:
pd.get_dummies(dataframe)

```python
# get_dummies aumenta o cardinalidade do dataset - 22 colunas
pd.get_dummies(dados.neighbourhood)
```

```python
# Para todas as variáveis o get_dummies aumentou em 13236 colunas - inviabiliza trabalhar com dataset
pd.get_dummies(dados)
```

## Aplicando o Label Encoder

No caso do Label Encoder podemos usar a indexação de variáveis:

 - No Python:
 
from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()

dataframe = lb_make.ﬁt_transform(dataframe)

 - Obs: É importante avaliar quando usar a técnica certa, categorizar de forma incorreta pode
enviesar o modelo

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le.fit_transform(dados['neighbourhood'])
```

```python
dados['nova_coluna'] = le.fit_transform(dados['neighbourhood'])
```

```python
pd.DataFrame(dados.dtypes).reset_index()
```

```python
colunas = pd.DataFrame(dados.dtypes).reset_index().rename(columns={'index': 'nome', 0: 'tipo'})
categ_cols = colunas[colunas.tipo == 'object']
```

```python
for i in categ_cols.index:
    dados[str(categ_cols['nome'][i]) + '_encoded'] = le.fit_transform(dados[categ_cols['nome'][i]])
    dados = dados.drop(categ_cols['nome'][i], axis = 1)
```
